```python
import xml.etree.ElementTree as ET

# XML dosyasının yolu
XML_FILE = "aek_data.xml"

# Kullanıcıdan AEK verilerini alma
def collect_aek_data():
    print("=== Açık Eğitim Kaynağı (AEK) Verilerini Girin ===")
    title = input("Başlık: ")
    subject = input("Anahtar Kelimeler (virgülle ayırın): ")
    description = input("Özet Bilgi: ")
    creator = input("Oluşturan Kişi/Kuruluş: ")
    license = input("Açık Lisans: ")
    language = input("Dil: ")
    identifier = input("Kaynağın URL'si: ")

    return {
        "Title": title,
        "Subject": subject,
        "Description": description,
        "Creator": creator,
        "License": license,
        "Language": language,
        "Identifier": identifier
    }

# XML'e veri ekleme
def save_to_xml(aek_data):
    try:
        tree = ET.parse(XML_FILE)
        root = tree.getroot()
    except FileNotFoundError:
        root = ET.Element("AEKList")

    aek = ET.SubElement(root, "AEK")
    for key, value in aek_data.items():
        element = ET.SubElement(aek, key)
        element.text = value

    tree = ET.ElementTree(root)
    with open(XML_FILE, "wb") as file:
        tree.write(file, encoding="utf-8", xml_declaration=True)

    print("Veri XML dosyasına kaydedildi.")

# XML'den verileri listeleme
def list_aek_data():
    try:
        tree = ET.parse(XML_FILE)
        root = tree.getroot()
        print("=== Kayıtlı AEK Verileri ===")
        for i, aek in enumerate(root.findall("AEK"), start=1):
            print(f"\nAEK {i}:")
            for element in aek:
                print(f"{element.tag}: {element.text}")
    except FileNotFoundError:
        print("XML dosyası bulunamadı. Henüz veri eklenmemiş.")

# XML'de arama ve filtreleme
def search_aek_data(keyword):
    try:
        tree = ET.parse(XML_FILE)
        root = tree.getroot()
        print(f"=== '{keyword}' Arama Sonuçları ===")
        found = False
        for aek in root.findall("AEK"):
            if any(keyword.lower() in (element.text or "").lower() for element in aek):
                found = True
                for element in aek:
                    print(f"{element.tag}: {element.text}")
                print("-" * 20)
        if not found:
            print("Arama kriterine uygun veri bulunamadı.")
    except FileNotFoundError:
        print("XML dosyası bulunamadı. Henüz veri eklenmemiş.")

# Ana menü
def main():
    while True:
        print("\n=== AEK Yönetim Sistemi ===")
        print("1. Veri Ekle")
        print("2. Verileri Listele")
        print("3. Veri Arama")
        print("4. Çıkış")
        choice = input("Seçiminizi yapın (1-4): ")

        if choice == "1":
            data = collect_aek_data()
            save_to_xml(data)
        elif choice == "2":
            list_aek_data()
        elif choice == "3":
            keyword = input("Arama için anahtar kelime girin: ")
            search_aek_data(keyword)
        elif choice == "4":
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçim! Lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()
```
